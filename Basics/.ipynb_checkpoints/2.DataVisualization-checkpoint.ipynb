```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

print("✅ Visualization libraries loaded!")
```

```python
x = np.linspace(0,10,50)
y = np.sin(x)

plt.plot(x, y, color='blue', marker='o', linestyle='--')
plt.title("Sine Wave Example")
plt.xlabel("X-axis")
plt.ylabel("Y = sin(x)")
plt.grid(True)
plt.show()
```

```python
students = ['Amit', 'Neha', 'Raj', 'Simran']
scores = [85, 90, 88, 92]

plt.bar(students, scores, color='orange')
plt.title("Student Scores")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

```python
# Create dataset
data = {
    'Student': ['Amit', 'Neha', 'Raj', 'Simran', 'Amit', 'Neha', 'Raj', 'Simran'],
    'Subject': ['Math', 'Math', 'Math', 'Math', 'Science', 'Science', 'Science', 'Science'],
    'Marks': [85, 90, 88, 92, 82, 95, 80, 87]
}
df = pd.DataFrame(data)

sns.barplot(x='Student', y='Marks', hue='Subject', data=df)
plt.title("Student Marks by Subject")
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.title("Restaurant Tips Analysis")
plt.show()
```
